```python
import pandas as pd
url = 'https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv'
food_data = pd.read_csv(url, encoding='latin1')

#13 total number and percentage of missing data in 2016 to 2 decimal places

food_data.isnull().sum()

#12 Mean and standard deviation for the year 2015

food_data.describe(include='all')

#11. Total sum of animal fat produced in 2014 and 2017

food_data.groupby('Item')[['Y2014','Y2017']].sum()
                            

food_data.describe

food_data.isnull().sum()

#15 The year with the highest sum of import quantity

food_data.groupby('Element')[['Y2014','Y2015','Y2016','Y2017','Y2018']].sum()

#16 Total number of the sum of production in 2014

food_data.groupby('Element')[['Y2014']].sum()

#17 The element having the highest sum in 2018

food_data.groupby('Element')[['Y2018']].sum().max()

#18 The element having the 3rd lowest sum in 2018

food_data.groupby('Element')[['Element','Y2018']].sum().min()

#19 Total import quantity in algeria

food_data.groupby('Area')[['Element', 'Y2018']].sum()

#20 total number of unique countries in the dataset

food_data.describe(include='all')
```
